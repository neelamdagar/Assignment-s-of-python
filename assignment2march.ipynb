```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# Ans: scatter plot help to find the relationship between two numerical value in the term of strength,  direction, sahpe, outlier.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("x -axis")
plt.ylabel("y - axis")
plt.title("scatter plot")
plt.scatter(x,y,c='r',alpha=0.9)
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
#subplot() function adds subplot to a current figure at the specified grid position.  
# It is similar to the subplots() function however unlike subplots() it adds one subplot at a time.
# One can use plt.subplots () to make all their subplots at once and it returns the figure and axes (plural of axis) of the subplots as a tuple
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
x2= np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([10, 20, 30, 40, 50, 60])
x3= np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.subplots(nrows=2,ncols=2)
for row in ax:
        for col in row:
            col.plot(x,y)
            col.plot(x1,y1)
            col.plot(x2,y2)
            col.plot(x3,y3)
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# answer : A bar plot or bar chart is a graph that represents the category of data with
# rectangular bars with lengths and heights that is proportional to the values which they represent.
# Bar plot is used to define the relationship between the one categorical and one numerical datatype.
plt.xlabel("x -axis")
plt.ylabel("y-axis")
plt.title("Bar plot")
plt.bar(company,profit)
           
```

```python
plt.barh(company,profit)
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# The method to summarize a set of data that is measured using an interval scale is called a box plot.
# it is used to ploting the only numerical datatype and show the exact value of data.
# we used quartile for ploting the box (i.e upper quartile, lower, meadian, (upper whisker and lower whisker used for to define outliers))
fig=plt.figure(figsize=(6,7))
plt.boxplot(box1)
```
