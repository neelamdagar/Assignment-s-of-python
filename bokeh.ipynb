```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
# # Display a variety of visual shapes whose attributes can be associated with data columns from ColumnDataSources.
# Any plot is usually made up of one or many geometrical shapes such as line, circle, rectangle, etc. These shapes have visual information about the corresponding set of data. In Bokeh terminology, these geometrical shapes are called gylphs.
```

```python
from bokeh.plotting import figure, output_file, show
xs = [[5,3,4], [2,4,3], [2,3,5,4]]
ys = [[6,4,2], [3,6,7], [2,4,7,8]]
fig = figure()
fig.patches(xs, ys, fill_color = ['red', 'blue', 'black'], line_color = 'white')
output_file('patch_plot.html')
show(fig)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
from bokeh.plotting import figure,output_file,show
x=[1,2,3,5]
y=[2,3,4,6]
output_file('line.html')
p=figure(title='test flower')

p.scatter(x,y,fill_color="red",legend_label="red points",size=10)
show(p)
```

```python
# Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
```

```python
from bokeh.plotting import figure
from bokeh.resources import CDN
from bokeh.embed import file_html

plot = figure()
plot.circle([1,2], [3,4])

html = file_html(plot, CDN, "my plot")
```
