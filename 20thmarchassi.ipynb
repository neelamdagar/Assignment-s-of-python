```python
import pandas as pd
import seaborn as sns
f=sns.load_dataset('iris')
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
k=encoder.fit_transform(f[['species']])
n=pd.DataFrame(k.toarray(), columns=encoder.get_feature_names_out())
r=pd.DataFrame(f['species'])
l=pd.concat([r,n],axis=1)
l
```
